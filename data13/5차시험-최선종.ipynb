```python
# 위 데이터프레임의 p컬럼으로 파이차트를 완성하시오.
plt.figure(figsize=(6,6))
df['p'].plot.pie(explode=[0.02, 0.02, 0.02, 0.02, 0.02], autopct='%1.1f%%')
```

```python
# x컬럼의 최대값을 구하시오.
df_x_max = df.x.max()
df_x_max
```

```python
# x컬럼의 최대값이 있는 y컬럼의 값을 출력
df_xequal_y = df[df.x == df_x_max]['y']
df_xequal_y
```

```python
# z컬럼을 그룹으로 묶어서 x값의 합계를 출력
df_z_xsum = df.groupby('z')['x'].sum()
df_z_xsum
```

```python
# z컬럼을 그룹으로 묶어서 p값의 평균을 출력 
df_z_pmean = df.groupby('z')['x'].mean()
df_z_pmean
```

```python
# p컬럼의 값에 모두 1을 더한 처리후, 데이터프레임을 출력
df_p_plus1 = []
for x in range(len(df['p'])):
    df['p'][x] = df['p'][x] + 1
    df_p_plus1.append(df['p'][x])
df_p_plus1
```

```python
df['p'] = df_p_plus1
df
```

```python
# z컬럼의 blue값을 모두 '파랑'으로 replace한 처리후, 데이터프레임을 출력
df['z'] = df['z'].replace('blue', '파랑')
df['z']
```
