```python
# raw_data['ì¹´í…Œê³ ë¦¬ëª…'], ì¹´ë ˆê³ ë¦¬ ì•ˆì˜ null ê°’ë“¤ ì œì™¸
raw_data = raw_data[raw_data['age'].notnull()]  
raw_data = raw_data[raw_data['sibsp'].notnull()]
raw_data = raw_data[raw_data['parch'].notnull()]
raw_data = raw_data[raw_data['fare'].notnull()]
raw_data.info()
```

```python
# í›ˆë ¨ ì‹œí‚¬ ì¹´í…Œê³ ë¦¬ë§Œ ì„ ë³„
train_pre = raw_data[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
train_pre
```

```python
# data ì¤€ë¹„ ë
# ê³µë¶€í•  ë°ì´í„°/ê²€ì¦ìš© ë°ì´í„° ë¶„ë¦¬
```

```python
# ë¨¸ì‹ ëŸ¬ë‹ìš© ë¼ì´ë¸ŒëŸ¬ë¦¬: sklearn, ì‚¬ì´í‚·ëŸ°
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_pre,  # ë°ì´í„°í”„ë ˆìž„
                 raw_data[['survived']],  # ë§žì¶”ì–´ì•¼í•  ì •ë‹µ
                 test_size = 0.1,  # í…ŒìŠ¤íŠ¸ë°ì´í„° 10% ì§€ì •, í›ˆë ¨ë°ì´í„° 90% í• ë‹¹
                 random_state = 13  # seedê°’                 
                )  
```

```python
len(X_train)  # í›ˆë ¨ìš© ë°ì´í„°ì˜ ë¬¸ì œ 
```

```python
len(y_train)  # í›ˆë ¨ìš© ë°ì´í„°ì˜ í•´ë‹µ
```

```python
X_train.head()
```

```python
X_train = X_train.reset_index()  # ì¸ë±ìŠ¤ ë¦¬ì…‹
X_train = X_train.drop(['index'], axis = 1)  # ì¸ë±ìŠ¤ê°’ dropì‹œí‚¤ê¸°

X_test = X_test.reset_index()
X_test = X_test.drop(['index'], axis = 1)

y_train = y_train.reset_index()
y_train = y_train.drop(['index'], axis = 1)

y_test = y_test.reset_index()
y_test = y_test.drop(['index'], axis = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier

# í•™ìŠµê¸° ê°ì²´ ìƒì„±
# ë§ˆë¼íŽ˜í„°ë¥¼ ë¶„ì„ê°€ê°€ ìž„ì˜ë¡œ ë³€ê²½í•  ìˆ˜ ìžˆìŒ.
# íŒŒë¼ë©”í„°ë¥¼ ë¶„ì„ê°€ê°€ ë³€ê²½ê°€ëŠ¥, ë³€ê²½ë¶ˆê°€ëŠ¥
# ë¶„ì„ê°€ê°€ ë³€ê²½ê°€ëŠ¥í•œ íŒŒë¼ë©”í„°ë¥¼ "í•˜ì´í¼íŒŒë¼ë©”í„°"ë¼ê³  í•¨

tree_clf = DecisionTreeClassifier(max_depth = 2, random_state = 13)  
# max_depth ëê¹Œì§€ ë“¤ì–´ê°€ì„œ í•™ìŠµí•˜ì§€ ì•Šê²Œ í•¨(ì»´í“¨í„°ëŠ” 100í”„ë¡œì¼ê²½ìš° ê·¸ê²ƒë„ ë¬¸ì œê°€ ë¨)
# random_state ë°ì´í„° ë¶„í• ì‹œ ì…”í”Œì´ ì´ë£¨ì–´ì§€ê³  ì´ë¥¼ ìœ„í•œ ì‹œë“œê°’

# ê³µë¶€ ì‹œí‚¤ê¸°
tree_clf.fit(X_train, y_train)

# trainë°ì´í„°ë¥¼ ê°€ì§€ê³  ê³µë¶€í•œ ì ìˆ˜
tree_clf.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score  # ì±„ì í•´ì£¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬

y_pred = tree_clf.predict(X_test)  # ë¬¸ì œë¥¼ ì£¼ê³  í’€ì–´ë³´ë¼ê³  í•¨. => ì •ë‹µ return! 

# í…ŒìŠ¤íŠ¸ë°ì´í„°ë¥¼ í‘¼ ë‹µì•ˆê³¼ í•´ë‹µì„ ë¹„êµí•´ì„œ ì •í™•ë„ë¥¼ êµ¬í•´ë³´ìž.
accuracy_score(y_test, y_pred)
```
