```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/COVID_Korea_geo.csv', encoding='euc-kr')
df
```

```python
df2 = df.dropna(axis=1)  # 열을 기준으로 Nan 값이 들어있는 열 삭제
df2
```

```python
df3 = df2.drop(['stdDay'], axis=1)
df3
```

```python
region = df3.sort_values(by='gubun', ascending=False).copy()
region
```

```python
gubun_region = region['gubun'] == '합계'
gubun_region
```

```python
total = region[gubun_region]
total
```

```python
total1 = total.sort_values(by='defCnt', ascending=False).copy()
total1
```

```python
covid_data = total1.iloc[0:11].copy()
covid_data
```

```python
covid_total = covid_data.drop(['gubun'], axis=1)
covid_total
```

```python
import seaborn as sns #시각화 추가 기능(색, 화면 나눔)
import matplotlib.pyplot as plt #시각화 가장 기본적인 기능
```

```python
sns.barplot(x='qurRate', y='deathCnt', data=covid_total)
```
